# **PreProcess**

```python
#importing necessary library
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("/content/OnlineRetail.csv")
```

```python
df.head()
```

```python
df['rate']=df['Quantity']*df['UnitPrice']
```

```python
df['InvoiceDate']=pd.to_datetime(df['InvoiceDate'])
```

```python
df['Date']=df['InvoiceDate'].dt.date
df['Time']=df['InvoiceDate'].dt.time
df['year']=df['InvoiceDate'].dt.year
df['month']=df['InvoiceDate'].dt.month
```

```python
df.head()
```

```python
df['HourFeature'] = df['Time'].astype(str).str.split(':').str[0].astype(int)
# Add 'AM' or 'PM' based on the hour
df['AMPM'] = df['HourFeature'].apply(lambda hour: 'AM' if hour < 12 else 'PM')
df['HourFeature'].drop
```

```python
print(len(df[df['Quantity']<=0]))
print(len(df[df['UnitPrice']<=0]))
df=df[df['Quantity']>0]
df=df[df['UnitPrice']>0]
print(len(df[df['Quantity']<=0]))
print(len(df[df['UnitPrice']<=0]))
```

# **Growth of customers from 2010 to 2011**


**Data may Collected or store may opened - From 2010-Dec to 2011-Dec**

```python
customer_count_per_year = df.groupby('year')['CustomerID'].nunique()

ax = customer_count_per_year.plot(kind='bar', xlabel='Year', ylabel='Number of Unique Customers', title='Customer Count per Year')

for i, count in enumerate(customer_count_per_year.values):
    ax.text(i, count + 0.1, str(count), ha='center')

plt.show()
```

# **Number Of Customer and Amount of Purchase Based On Country**

```python
customer_count_per_country = df.groupby('Country')['CustomerID'].nunique()


fig, ax = plt.subplots(figsize=(14, 6))


ax.bar(customer_count_per_country.index, customer_count_per_country.values)
ax.set_xlabel('Country')
ax.set_ylabel('Number of Unique Customers')
ax.set_title('Customer Count per Country')

# Display counts on the bars
for i, count in enumerate(customer_count_per_country.values):
    ax.text(i, count + 0.1, str(count), ha='center')

ax.tick_params(axis='x', rotation=90)
plt.tight_layout()
plt.show()
```

```python
customer_count_per_country = df.groupby('Country')['rate'].sum()


fig, ax = plt.subplots(figsize=(14, 6))


ax.bar(customer_count_per_country.index, customer_count_per_country.values)
ax.set_xlabel('Country')
ax.set_ylabel('Number of Amount of Purchase')
ax.set_title('Purchase Amount per Country')

# Display counts on the bars
for i, count in enumerate(customer_count_per_country.values):
    ax.text(i, count + 0.1, str(int(count)), rotation=90,ha='center')

ax.tick_params(axis='x', rotation=90)
plt.tight_layout()
plt.show()
```

# **TIMELINE BASED CHARTS**

```python
feature_counts = df['AMPM'].value_counts()

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(feature_counts, labels=feature_counts.index, autopct='%1.1f%%', startangle=90, colors=['lightblue', 'lightcoral'])
plt.title('Distribution of Feature1 (AM/PM)')
```

**Most Of The User Buy Product At PM**

```python
hourly_counts = df['HourFeature'].value_counts().sort_index()

# Create a line chart
plt.figure(figsize=(10, 6))
plt.plot(hourly_counts.index, hourly_counts.values, marker='o', linestyle='-', color='b')
plt.title('Hourly Customer Counts')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Customers')
plt.grid(True)

# Show the line chart
plt.show()
```

*   **peak hour of the store is 12:00 to 14:30 o'clocks**
*   **20:00 to 6:00 maybe the shop closing time**
